```python
query_model = pipeline("text2text-generation", model="google/flan-t5-large")
```

```python
from google.colab import files

uploaded = files.upload()  # User uploads a CSV file

file_name = list(uploaded.keys())[0]  # Get the file name
df = pd.read_csv(file_name)  # Read the CSV file

print("Data Preview:")
print(df.head())  # Show first few rows
```

```python
def query_data(query, df):
    # Convert DataFrame to a text format for AI model processing
    data_summary = df.describe(include="all").to_string()  # Get summary stats
    input_text = f"Data:\n{data_summary}\n\nQuestion: {query}\nAnswer:"

    # Get AI response
    response = query_model(input_text, max_length=200, truncation=True)[0]["generated_text"]
    return response
```

```python
user_query = input("Enter your query about the data: ")
response = query_data(user_query, df)

print("\nAI Response:")
print(response)
```
